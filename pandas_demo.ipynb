```python
# Finding how many percentage of people know python from all countries
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'No_of_respondents', 'LanguageWorkedWith': 'People_who_know_python'}, inplace=True)
```

```python
python_df
```

```python
python_df['PercentageKnowPython'] = (python_df['People_who_know_python'] / python_df['No_of_respondents']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by='PercentageKnowPython', ascending=False, inplace=True)
python_df
```
